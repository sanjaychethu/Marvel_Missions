```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime


data = {
    'Date': pd.date_range(start='2023-01-01', end='2023-12-31', freq='M'),
    'Launches': [5, 3, 7, 2, 8, 6, 4, 9, 3, 5, 7, 6]
}

df = pd.DataFrame(data)

# Line Graph
plt.figure(figsize=(10, 5))
plt.plot(df['Date'], df['Launches'], marker='o', color='b', label='Monthly Launches')
plt.title('Monthly ISRO Launches Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Launches')
plt.legend()
plt.grid(True)
plt.show()

# Bar Graph
plt.figure(figsize=(10, 5))
plt.bar(df['Date'], df['Launches'], color='g', alpha=0.7, label='Monthly Launches')
plt.title('Monthly ISRO Launches Bar Graph')
plt.xlabel('Date')
plt.ylabel('Number of Launches')
plt.legend()
plt.grid(axis='y')
plt.show()

# Scatter Plot
plt.figure(figsize=(10, 5))
plt.scatter(df['Date'], df['Launches'], color='r', marker='x', label='Monthly Launches')
plt.title('Scatter Plot of Monthly ISRO Launches')
plt.xlabel('Date')
plt.ylabel('Number of Launches')
plt.legend()
plt.grid(True)
plt.show()
```
